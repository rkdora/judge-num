```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title('LogisticRegression', fontsize = 20)
x = np.array([500, 5000, 50000])
logi = np.array([120, 3400, 7200])
plt.plot(x, logi)
plt.show()
```

```python
plt.title('SVM', fontsize = 20)
svm = np.array([310, 1300, 535000])
plt.plot(x, svm)
plt.show()
```

```python
plt.plot(x, logi)
plt.plot(x, svm)
plt.show()
```
